```python
import pandas as pd
import requests
```

```python
# Processo de importação dos dados
df_train = pd.read_csv('train.csv',delimiter=',')
df_train.head()
```

```python
nomes_columas = list(df_train.columns)
nomes_columas = [i.lower() for i in nomes_columas]
nomes_columas[0]='id'
df_train.columns = nomes_columas
df_train.head()
```

```python
# Ver quantidade de valores faltantes:
df_train.isna().sum()
```

```python
# Ver Estatísticas dos dados numéricos:
df_train.describe()
```

```python
# No caso da idade, os valores faltantes serão substituidos pela mediana.
df_train['age'] = df_train['age'].fillna(df_train['age'].median())
df_train.describe()
```

```python
df_train = df_train.drop(columns=['ticket','cabin'])
df_train
```

```python
import json
nomes_columas_list = list(df_train.columns)
nomes_columas_list
```

```python
import sqlite3
conec = sqlite3.connect('db.sqlite3')
```

```python
cursor = conec.cursor()
```

```python
cursor.execute(
    """
    SELECT * FROM apiformstitanic_base
    """
)
for linha in cursor.fetchall():
    print(linha)
```

```python
df_train['name'] = df_train['name'].str.split(expand=True)[0]
df_train['name'] = df_train['name'].str.title()
df_train['name'] = df_train['name'].str.replace(',','')

df_train
```

```python
for i in range(len(df_train)):
    id_ = int(df_train['id'][i])+1
    survived = int(df_train['survived'][i])
    pclass = int(df_train['pclass'][i])
    sex = str(df_train['sex'][i]).strip()
    age = int(df_train['age'][i])
    name__ = str(df_train['name'][i]).strip()
    sibsp = int(df_train['sibsp'][i])
    parch = int(df_train['parch'][i])
    fare = float(df_train['fare'][i])
    embarked = str(df_train['embarked'][i]).strip()

    cursor.execute(
        f"""INSERT INTO apiformstitanic_base 
        (id,survived, pclass,age, name, sibsp, parch, fare, embarked, sex )
        VALUES 
    ({id_},{survived},{pclass},{age},"{name__}",{sibsp},{parch},{fare},"{embarked}","{sex}");"""
    )
```

```python
conec.commit()
```
